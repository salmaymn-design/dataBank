```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("postgresql://postgres:password@localhost/financecore_db")
```

```python
df = pd.read_csv("financecore_clean.csv")

df.head()
```

```python
clients = df[['client_id', 'segment_client', 'score_credit_client']].drop_duplicates()
segments = df[['segment_client']].drop_duplicates()
agences = df[['agence']].drop_duplicates()
produits = df[['produit']].drop_duplicates()
transactions = df[['transaction_id', 'client_id', 'produit', 'montant', 'date_transaction']]
```

```python
segments.to_sql("segments", engine, if_exists='append', index=False)
clients.to_sql("clients", engine, if_exists='append', index=False)
agences.to_sql("agences", engine, if_exists='append', index=False)
produits.to_sql("produits", engine, if_exists='append', index=False)
```
